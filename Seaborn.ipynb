### Dirstibution plot

- Distplot
- Jointplot
- pairplot

```python
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
 df = sns.load_dataset('tips')
```

```python
df.head()
```

### Correlation with  heatmap

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

### Joint plot

```python
sns.jointplot(x='tip', y='total_bill', data = df)
```

```python
sns.jointplot(x='tip', y='total_bill', data = df, kind='hex')
```

```python
sns.jointplot(x='tip', y='total_bill', data = df, kind='reg')
```

```python
sns.jointplot(x='tip', y='total_bill', data = df, kind='kde')
```

### Pairplot

```python
sns.pairplot(df) #Feature need to be int/float, also called as scatter plot in pyplot
```

```python
sns.pairplot(df, hue='sex')
```

```python
sns.pairplot(df, hue='smoker')
```

### Dist plot

```python
#Histogram 

sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'], kde=False, bins=10) #kde - kernel density estimation
```

## Categorical plots
- boxplot
- violinplot
- countplot
- bar plot

```python
#count plot
sns.countplot('sex', data=df)
```

```python
sns.countplot('smoker', data=df)
```

```python
sns.countplot('day', data=df)
```

```python
sns.countplot(y='day', data=df)
```

### Bar plot

```python
sns.barplot(x='total_bill', y='day', data=df)
```

```python
sns.barplot(y='total_bill', x='day', data=df)
```

```python
sns.barplot(y='total_bill', x='sex', data=df)
```

```python
sns.barplot(y='total_bill', x='smoker', data=df)
```

### Box plot

```python
sns.boxplot('sex','total_bill', data=df)
```

```python
sns.boxplot('total_bill', data=df)
```

```python
sns.boxplot(x='total_bill',y='day', data=df, palette='rainbow',orient='h')
```

```python
sns.boxplot(x='total_bill',y='day', data=df, hue='smoker')
```

### Violin plot

- Violin plot helps us to  see both the distribution of data in kernel density estimation and box plot

```python
sns.violinplot(x='total_bill', y='day', data=df, palette='rainbow')
```
